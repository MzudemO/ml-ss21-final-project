```python
import matplotlib.pyplot as plt
import json
```

```python
hst = []
with open("history_lr0.01_sgd_train_full_scratch.json", "r") as f:
    hst = json.load(f)

plt.plot(hst["loss"], label="Training loss")
plt.plot(hst["val_loss"], label="Validation loss")
plt.legend()
plt.title("Learning from scratch")
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.show()
```
